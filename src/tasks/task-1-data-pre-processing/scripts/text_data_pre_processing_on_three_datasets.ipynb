## Part 1 : Hate speech and offensive language dataset text preprocessing

```python
df = pd.read_csv('labeled_data.csv',index_col=[0])
df1 = pd.read_csv('train_E6oV3lV.csv')
df.head()
```

knowing more about data

```python
df.info()
```

mapping the class name according to labels and creating a new column name as category

```python
c=df['class']
df.rename(columns={'tweet' : 'text',
                   'class' : 'category'}, 
                    inplace=True)
a=df['text']
b=df['category'].map({0: 'hate_speech', 1: 'offensive',2: 'neither'})

df= pd.concat([a,b,c], axis=1)
df.rename(columns={'class' : 'label'}, 
                    inplace=True)
```

```python
df.head()
```

removing rt (retweet) from each sentence

```python
import re
df['text'] = df['text'].apply(lambda x : re.sub(r'\bRT\b','', x).strip())
```

removing usernames from each sentence

```python
df['text'] = df['text'].apply(lambda x : re.sub(r'@[^\s]+','', x).strip())
```

```python
df.head()
```

removing urls using texthero library

```python
df['text'] = hero.remove_urls(df['text'])
```


>performing text cleaning by removing:

*   digits

*   punctuation

*   diacritics

*   whitespace











```python
df['text'] = hero.clean(df['text'])
```

```python
df.head()
```

```python
df.drop('label', axis =1, inplace=True)
```

```python
df['category'].value_counts()
```

```python
df.to_csv('hate_speech6.csv')
```

##Part 5: Lemmatizing the combined dataset for model building

```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
df5 = pd.read_csv('combined_dataset.csv')
df5.head()
```

```python
df5['category'].value_counts()
```

```python
df5['text'] = df5['text'].apply(str)# map dataframe to string
```

removing stopwords

```python
df5['text'] = hero.preprocessing.remove_stopwords(df5['text'])
```

applying Lemmatizing on this dataset

```python
import nltk # import nltk library for lemmatization
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
def lemmatize_words(text): # creating function to lemmatize each sentence from dataframe
    words = text.split() # spliting sentence into words before lemmatiztion
    words = [lemmatizer.lemmatize(word,pos='v') for word in words] # apply lemmatiztion on each word by loop
    return ' '.join(words) # joining all lemmatize words into sentences
df5['text'] = df5['text'].apply(lemmatize_words) # applying the function
```
